```python
# Importing the required libraries
import pandas as pd
import numpy as np
import re

pd.set_option('display.max_columns', None)
# pd.set_option('display.max_rows', None)
```

```python
# Importing the raw dataset
df = pd.read_csv('DataAnalyst.csv')
df.drop('Unnamed: 0', axis = 1, inplace = True)
```

```python
def get_a_b(x):
    a = []
    b = []
    for row in x:
        a.append(row[0])
        b.append(row[1])
    return a,b
```

```python
def get_min_max_salary(salary_estimate):
    pattern = '[^\d-]'
    min_, max_ = re.sub(pattern, '', salary_estimate).split('-')
    return min_,max_
```

```python
# Cleaning the data column wise

df['Min Salary Estimate'], df['Max Salary Estimate'] = get_a_b(df['Salary Estimate'].apply(get_min_max_salary))
df['Min Salary Estimate']  =  pd.to_numeric( df['Min Salary Estimate']) * 1000
df['Max Salary Estimate']  =  pd.to_numeric( df['Max Salary Estimate']) * 1000
df['Company Name'] = df['Company Name'].str.split('\n').str[0]
df['City'] = df['Location'].str.split(',').str[0]
df['State'] = df['Location'].str.split(',').str[1].str.strip()
df['HCity'] = df['Headquarters'].str.split(',').str[0]
df['HState'] = df['Headquarters'].str.split(',').str[1].str.strip()

df.replace(['-1', 'Unknown'], np.nan, inplace = True)

size_dict = {"1 to 50 employees":'1-50',
            "51 to 200 employees":'51-200',
            "201 to 500 employees":'201-500',
            "501 to 1000 employees":'501-1000',
            "1001 to 5000 employees":'1001-5000',
            "5001 to 10000 employees":'5001-10000',
            "10000+ employees":'10000+'}
df["Size"] = df["Size"].apply(lambda x: size_dict.get(x))

df.replace({'Type of ownership': {'Company - Private': 'Private Company',
                                 'Company - Public': 'Public Company',
                                 'Subsidiary or Business Segment': 'Business',
                                 'College / University': 'Education',
                                 'Other Organization': np.nan,
                                 'Private Practice / Firm': 'Private Firm',
                                 'School / School District': 'Education'}}, inplace=True)

df.replace({'Revenue': {'Unknown / Non-Applicable': np.nan}}, inplace=True)
df['Revenue'] = df['Revenue'].str.replace('$','').str[:-5].str.strip()

df['Number of Competitors'] = df['Competitors'].apply(lambda x: len(x.split(',')) if not isinstance(x, float) else np.nan)
```

```python
columns = ['Job Title',
          'Min Salary Estimate', 
          'Max Salary Estimate',
          'Rating',
          'Company Name',
          'City',
          'State', 
          'HCity',
          'HState',
          'Size',
          'Founded',
          'Type of ownership',
          'Industry',
          'Sector', 
          'Revenue',
          'Number of Competitors',
          'Easy Apply']
```

```python
# Storing the cleaned data in a csv file
df = df[columns]
df.to_csv('DataAnalystCleaned.csv', index=True)
```
